```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isnull().sum()
```

```python
# TODO: drop your null values
coffee.dropna()
```

```python
# TODO: Select only Minas Gerais

minas_gerais = coffee[coffee["subdivision"] == "Minas Gerais"]
minas_gerais.head(10)
```

```python
# TODO: BEGIN YOUR EDA!

# bearing trees over the years
sns.lineplot(minas_gerais, x="year", y="bearing_trees")
plt.show()
```

```python
#distribution of bearing trees
sns.histplot(minas_gerais, x="bearing_trees", kde=True)
plt.show()
```

```python
minas_gerais.describe()
```

```python
# bearing hectares over the years
sns.lineplot(minas_gerais, x="year", y="bearing_hectares")
plt.show()
```

```python
# distribution of bearing hectares

sns.histplot(minas_gerais, x="bearing_hectares", kde=True, bins=10)
plt.show()
```

```python
# change in count per million 60kg coffee bags
sns.lineplot(minas_gerais, x="year", y="60kgs_bag")
plt.show()
```

```python
# distribution of 60kg bags

sns.histplot(minas_gerais, x=minas_gerais["60kgs_bag"], kde=True, bins=10)
plt.show()
```

```python
# relation between bearing trees and 60kg bags coffee
coffee[["bearing_trees", "60kgs_bag"]].corr()
```

```python
# scatterplot for bearing trees and 60kg bags
sns.scatterplot(coffee, x="bearing_trees", y="60kgs_bag")
plt.show()
```

```python
# relation between bearing_hectares and 60kgs_bag
coffee[["bearing_hectares", "60kgs_bag"]].corr()
```

```python
# scatterplot for bearing_hectares and 60kgs_bag
sns.scatterplot(coffee, x="bearing_hectares", y="60kgs_bag")
plt.show()
```
